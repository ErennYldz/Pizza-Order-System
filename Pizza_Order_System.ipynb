```python
import csv
import datetime
import locale

def main():
    # Menüyü ekrana yazdırıyoruz.
    f = open("menu.txt", "r", encoding="utf8")
    print(f.read())
    
    # Pizzaları txt dosyasında yazan numaralarına göre seçiyoruz.
    print("Sipariş edeceğiniz pizzanın kodunu girin.")
    x = input()
    #Pizza tabanlarının sahip olduğu özellikleri belirtiyoruz.
    class pizzatür:
        def __init__(self, isim, fiyat, hamur):
            self.isim = isim
            self.fiyat = fiyat
            self.hamur = hamur 
            
    # Seçilen koda göre pizzanın özellikleri belirleniyor.        
    if int(x) == 1:
        tür = pizzatür("Klasik Pizza", 40, "normal")
    elif int(x) == 2:
        tür = pizzatür("Margarita Pizza", 50, "ince")
    elif int(x) == 3:
        tür = pizzatür("Türk Pizza", 60, "kalın")
    elif int(x) == 4:
        tür = pizzatür("Sade Pizza", 30, "normal")
    else:
        return("Geçersiz Pizza Kodu!")
        
    # Pizzanın tanımını yapıp fiyatını belirtiyoruz.
    class pizza:
        def get_cost():
            print(f"{tür.isim},{tür.fiyat} türk lirasıdır.")
        def get_description():
            print(f"{tür.isim},{tür.hamur} hamurlu bir pizzadır.")   
            
    pizza.get_description()
    pizza.get_cost()
    
    # Sosları txt dosyasında yazan numaralarına göre seçiyoruz.
    print("Eklemek istediğiniz sosun kodunu girin.")
    y = input()
    
    #Pizza soslarının sahip olduğu özellikleri belirtiyoruz.
    class soslar:
        def __init__(self, isim, fiyat, tanım):
            self.isim = isim
            self.fiyat = fiyat
            self.tanım = tanım
            
     # Seçilen koda göre sosun özellikleri belirleniyor.
    if int(y) == 11:
        sos = soslar("Zeytin", 2, "zeytinli")
    elif int(y) == 12:
        sos = soslar("Mantar", 3, "mantarlı")
    elif int(y) == 13:
        sos = soslar("Keçi Peyniri", 4, "keçi peynirli")
    elif int(y) == 14:
        sos = soslar("Et", 5, "etli")
    elif int(y) == 15:
        sos = soslar("Soğan", 2, "soğanlı")
    elif int(y) == 16:
        sos = soslar("Mısır", 2, "mısırlı")
    else:
        return("Geçersiz Sos Kodu!")
        
     # Sosun fiyatını belirtiyoruz
    def get_cost_sos():
        print(f"{sos.isim},{sos.fiyat} türk lirasıdır.")   
    
    get_cost_sos()
    
    #Toplam ücreti hesaplıyoruz.
    toplam_ücret = tür.fiyat + sos.fiyat
    print(f"Seçtiğiniz pizza {sos.tanım} {tür.isim}'dır")
    print(f"Toplam ücret {toplam_ücret} türk lirasıdır.")
    print("Ödeme için bilgilerinizi giriniz.")
    
    #Ödeme için gerekli olan kullanıcı bilgilerini alıyoruz.
    kullanıcı_adı = input("İsim-Soyisim :")
    tc = input("TC kimlik numaranısı: ")
    kredi_kart_numarası = input("Kart Numarası: ")
    kart_şifresi = input("Kartın şifresi: ")
    print(f"Tebrikler, {sos.tanım} {tür.isim} siparişiniz başarıyla alınmıştır!")
    
    #İşlem saatini öğreniyoruz.
    locale.setlocale(locale.LC_ALL, '')
    an = datetime.datetime.now()
    tarih = datetime.datetime.strftime(an, '%c')
    print(f"İşlem Zamanı: {tarih}")
    
    #Kullanıcı bilgilerini veritabanına işliyoruz.
    with open('Orders_Database.csv', mode='w') as database:
        customer_data = csv.writer(database, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    
        customer_data.writerow(["İsim-Soyisim", "TC Kimlik Numarası", "Kredi Kart Numarası", "Kart Şifresi","Sipariş Türü", "İşlem Tarihi"])
        customer_data.writerow([kullanıcı_adı, tc, kredi_kart_numarası, kart_şifresi,tür.isim, tarih])
    
main()
```
